```python
import zipfile
import pandas as pd
import os

# 1) Define paths
zip_path = "C:/Users/afric/Downloads/census+income (1).zip"
extract_dir = '/mnt/data/census_income_files'

# 2) Extract all files from the zip
os.makedirs(extract_dir, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_dir)

# 3) Define column names (as per the “Adult” census dataset)
col_names = [
    "age", "workclass", "fnlwgt", "education", "education_num",
    "marital_status", "occupation", "relationship", "race", "sex",
    "capital_gain", "capital_loss", "hours_per_week", "native_country", "income"
]

# 4) Build paths to the train/test files
adult_data_path = os.path.join(extract_dir, 'adult.data')
adult_test_path = os.path.join(extract_dir, 'adult.test')

# 5) Read the training set
df_train = pd.read_csv(
    adult_data_path,
    header=None,
    names=col_names,
    sep=r',\s*',      # comma + optional whitespace separator
    engine='python'
)

# 6) Read the test set (skip the first row, which is a header/comment)
df_test = pd.read_csv(
    adult_test_path,
    header=None,
    names=col_names,
    sep=r',\s*',
    engine='python',
    skiprows=1
)

# 7) Concatenate train and test into one DataFrame
combined_df = pd.concat([df_train, df_test], ignore_index=True)

# 8) Inspect the result
print("Combined shape:", combined_df.shape)
print(combined_df.head())
```
